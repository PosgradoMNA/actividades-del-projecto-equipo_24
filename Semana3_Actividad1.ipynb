$\;\;\;\;\;$**1.- Fundamentos de bases de datos y para ciencia de datos.**

El procesamiento de los datos es una parte fundamental para que estos puedan ser utlizados en modelos. La razón es qeu estos pueden estar incompletos, no representar los atributos, ser inconsistentes o presentar ruido. Cuando se habla del procesamiento de datos, se incluye de manera global: Limpieza de datos, transformación de datos y reducción de datos. En otras palabras, se puede decir que es el proceso de identificar y corregir las entradas erróneas.

Se pueden identificar de manera general tres tipos de datos faltantes:
 - **Faltantes Completamente Aleatorios (*MCAR*, Missing Completely at Random)**: No se relaciona coon el valor que se supone que se obtienen, ni con las respuestas observadas. Cuentan como MCAR si se pierden en el tránsito o son insatisfactorias.
 - **Faltantes Aleatorios (*MAR*, Missing at Random)**: Depende del conjunto de respuestas observadas, pero no está relacionada con los valores faltantes que se supone se obtienen.
 - **Faltantes No Aleatorios (*MNAR*, Missing Not at Random)**: Los datos faltantes no son MCAR ni MAR.
 
Para el manejo de estos datos faltantes se tienen las siguientes técnicas, por mencionar algunas:
 - **Eliminación por lista**: Consiste en eliminar todas las instancias que tengan datos faltantes. Esto es una aproximación recomendable si se cuenta con una muestra lo suficientemente grande.
 - **Eliminación por parejas**: Consiste en eliminar una instancia sólo si esta es parte de el análisis a realizar. Se utlizan los datos existentes para las pruebas estadísticas.
 - **Sustitución media**: Los datos faltantes se reemplazan por el valor medio de esa variable. Recomendable en una distribución normal.
 - **Sustitución de la mediana**: Los datos faltantes se reeemplazan por la mediana de la misma variable. Recomendable si los datos faltantes son estrictamente aleatorios.
 - **Imputación de Regresión**: Reemplza el dato faltante con un valor probable estimado usando la información disponible. Esta aproximación no agrega información novedosa y altera ciertos valores estadísticos como  el error estándar. Crece el tamaño de la muesta.
 - **Imputación Múltiple**: Se predicen valores con base en los datos que sí se tienen y se crea un conjunto de datos con estas propuestas. Incorpora incertidumbre a los valores pronosticados y puede reemplazar los valores faltantes con los imputados.
 - **Imputación de cubierta caliente**: Reemplazar el valor faltante de la instancia por uno escogido de maneraaleatoria por de otra instancia con caractarísitcas conocidas similares.
 - **Imputación de cubierta fría**: Reemplazar el valor faltante de la instancia por otra instancia concaractarísitcas conocidas similares. Similar a imputación de cubierta caliente, pero sin que sea aleatorio.
 - **Imputación de regresión escolástica**: Reemplazar el valor faltante por una regresión adicionandoun valor residual.
 
Cada forma de tratar los datos faltantes tiene sus ventajas y desventajas. Es importante tenerlas en cuenta cuando se propone una u otra forma de manejar estos datos. 

$\;\;\;\;\;$**2.- Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.**

Un Almacén de Datos (Data Warehouse) es un conjunto de bases de datos gigantesca, la cual junta información de distintas fuentes con el fin de poder realizar análisis. Estas bases de datos pueden almacenar información de distintos tipos en cuestión de formatos, pero también en cuestión del origen de los mismos: históricos, en tiempo real o procesados.

Dentro de las tareas que tiene un Data Warehouse se incluye la colocación de los datos en un esquema (schema) que permita el análisis en simultáneo. Para esto se manejan comunmente dos corrientes de procesamiento de los datos: ETL y ELT.

ETL proviene de Extract, Transform, Load. Esto quiere decir que los datos primero se extraen de su origen, posteriormente se transforman para ser congurentes y, por último, se vuelven disponibles como parte del Warehouse. Por otro lado, ELT proviene de Extract, Load, Transfomr. En este sentido, se saca provecho de el cómputo en la nube de tal forma que primero se suben los datos como se obtuvieron y ya en el espacio donde serán almacenados, se procesan para ser congruentes.

Hablando de herramientas puntuales, se menciona SQL como una de las principales para poderse conectar y editar bases de datos, ya sea desde un entorno gáfico o a través de queries. Las bases de datos como colecciones de tablas relacionadas entre sí, siguen una descripción dada por el 'schema', teniendo como posibles relaciones: One-to-many, Many-to-many y One-to-one.

Antes de empezar a trabajar con datos, es importante conocer el origne de los mismos, así como ciertos aspectos clave como lo son:
 - Los datos relevantes
 - Identificadores de las tablas
 - Si los datos fueron previamente procesados
 - Si la tabla recibe alguna especia de actualización
 
Por último, es importante mencionar la diferencia que existe entre Data Warehouse y Data Lake. El Data Warehouse contempla el procesamiento de los datos, mientras que el Data Lake se concentra sólo en el almacenamiento de los datos crudos.

##### Solución 2.1: Sustituir por la media

```python
df2_1 = df.copy()
for column in df2_1.columns:
    avg = df2_1[column].mean()
    df2_1[column].fillna(value = avg, inplace = True)
```

```python
df2_1.isnull().values.any()
```

```python
# df2_1.describe()
```

Esta sustitución puede generar problemas en valores estadísticos como el error o la desviación estándar.

<a id='Sol2_2'></a>
##### Solución 2.2: Sustituir por la mediana

```python
df2_2 = df.copy()
for column in df2_1.columns:
    med = df2_2[column].median()
    df2_2[column].fillna(value = med, inplace = True)
```

```python
df2_2.isnull().values.any()
```

```python
# df2_2.describe()
```

Esta sustitución puede generar problemas si los datos tienen una distribución uniforme.

<a id='Sol2_3'></a>
##### Solución 2.3: Sustituir por la moda

```python
df2_3 = df.copy()
for column in df2_3.columns:
    mod = df2_3[column].mode()[0]
    df2_3[column].fillna(value = mod, inplace = True)
```

```python
df2_3.isnull().values.any()
```

```python
# df2_3.describe()
```

Esta sustitución puede no ser conveniente en casos donde los datos no sean categóricos.

<a id='Sol2_4'></a>
##### Solución 2.4: Sustituir por método de preferencia

```python
df2_4 = df.copy()

favs = {'X1': df2_4['X1'].mean(),		# Monto del crédito dado
        'X2': df2_4['X2'].mode()[0],    # Género
        'X3': df2_4['X3'].mode()[0],    # Nivel de educación
        'X4': df2_4['X4'].mode()[0],    # Estado civil
        'X5': df2_4['X5'].mean(),       # Edad
        'X6': df2_4['X6'].mode()[0],    # Status de pago septiembre
        'X7': df2_4['X7'].mode()[0],    # Status de pago agosto
        'X8': df2_4['X8'].mode()[0],    # Status de pago julio
        'X9': df2_4['X9'].mode()[0],    # Status de pago junio
        'X10': df2_4['X10'].mode()[0],  # Status de pago mayo
        'X11': df2_4['X11'].mode()[0],  # Status de pago abril
        'X12': df2_4['X12'].mean(),     # Monto del estado de cuenta septiembre
        'X13': df2_4['X13'].mean(),     # Monto del estado de cuenta agosto
        'X14': df2_4['X14'].mean(),     # Monto del estado de cuenta julio
        'X15': df2_4['X15'].mean(),     # Monto del estado de cuenta junio
        'X16': df2_4['X16'].mean(),     # Monto del estado de cuenta mayo
        'X17': df2_4['X17'].mean(),     # Monto del estado de cuenta abril
        'X18': df2_4['X18'].mean(),     # Monto del pago del mes anterior a septiembre
        'X19': df2_4['X19'].mean(),     # Monto del pago del mes anterior a agosto
        'X20': df2_4['X20'].mean(),     # Monto del pago del mes anterior a julio
        'X21': df2_4['X21'].mean(),     # Monto del pago del mes anterior a junio
        'X22': df2_4['X22'].mean(),     # Monto del pago del mes anterior a mayo
        'X23': df2_4['X23'].mean(),     # Monto del pago del mes anterior a abril
        'Y': df2_4['Y'].mode()[0]       # Variable de salida: Pago por default
       }

for column in df2_4.columns:
    if column in favs:
        df2_4[column].fillna(value = favs[column], inplace = True)
```

```python
df2_4.isnull().values.any()
```

```python
# df2_4.describe()
```

Esta sustitución puede ser personalizada para cada columna, minimizando así el impacto negativo de los distintos tipos de sustituciones.

Se realizará primero una eliminación con threshold donde se eliminen aquellas filas que no contengan al menos 15 valores en ellos.

```python
df3_1 = df.copy()
df3_1.dropna(thresh=15, inplace = True)
```

Posteriormente se realizará la sustitución por media, moda y mediana, según sea el caso. Para esto se consideró que valores continuos se sustituyeran por la media, mientra que los categóricos se sustituyeran por la moda.

```python
favs = {'X1': df3_1['X1'].mean(),		# Monto del crédito dado
        'X2': df3_1['X2'].mode()[0],    # Género
        'X3': df3_1['X3'].mode()[0],    # Nivel de educación
        'X4': df3_1['X4'].mode()[0],    # Estado civil
        'X5': df3_1['X5'].mean(),       # Edad
        'X6': df3_1['X6'].mode()[0],    # Status de pago septiembre
        'X7': df3_1['X7'].mode()[0],    # Status de pago agosto
        'X8': df3_1['X8'].mode()[0],    # Status de pago julio
        'X9': df3_1['X9'].mode()[0],    # Status de pago junio
        'X10': df3_1['X10'].mode()[0],  # Status de pago mayo
        'X11': df3_1['X11'].mode()[0],  # Status de pago abril
        'X12': df3_1['X12'].mean(),     # Monto del estado de cuenta septiembre
        'X13': df3_1['X13'].mean(),     # Monto del estado de cuenta agosto
        'X14': df3_1['X14'].mean(),     # Monto del estado de cuenta julio
        'X15': df3_1['X15'].mean(),     # Monto del estado de cuenta junio
        'X16': df3_1['X16'].mean(),     # Monto del estado de cuenta mayo
        'X17': df3_1['X17'].mean(),     # Monto del estado de cuenta abril
        'X18': df3_1['X18'].mean(),     # Monto del pago del mes anterior a septiembre
        'X19': df3_1['X19'].mean(),     # Monto del pago del mes anterior a agosto
        'X20': df3_1['X20'].mean(),     # Monto del pago del mes anterior a julio
        'X21': df3_1['X21'].mean(),     # Monto del pago del mes anterior a junio
        'X22': df3_1['X22'].mean(),     # Monto del pago del mes anterior a mayo
        'X23': df3_1['X23'].mean(),     # Monto del pago del mes anterior a abril
        'Y': df3_1['Y'].mode()[0]       # Variable de salida: Pago por default
       }

for column in df3_1.columns:
    if column in favs:
        df3_1[column].fillna(value = favs[column], inplace = True)
```

Esto hace que ya no se tengan valores faltantes
