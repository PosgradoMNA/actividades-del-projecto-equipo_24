Se trata de una colección de KPI's mantenida por OWD que incluye:
 - Datos anuales, per capita, acumulativos y basados en consumo de las emisiones de CO2.
 - Emisiones de otros gases de invernadero.
 - Mézcla de energias.

La documentación y los datasets se pueden encontrar en el siguiente _[enlace](https://github.com/owid/co2-data)_ de GitHub. Los datos se pueden obtener en archivos CSV, XLSX y JSON.

El dataset es el resultado de la concentración y transformación de otras fuentes que se enumerarán a continuación (si se desea ahondar en los ETL's de cada fuente, se pueden consultar en la documentación del dataset co2-data):

 - Statistical review of world energy (BP): _[Datos origen](https://www.bp.com/en/global/corporate/energy-economics/statistical-review-of-world-energy.html)_
 - International energy data (EIA): _[Datos origen](https://www.eia.gov/opendata/bulkfiles.php)_
 - Primary energy consumption (Our World in Data based on BP's Statistical review of world energy & EIA's International energy data)
 - Global carbon budget (Our World in Data based on Global Carbon Project): _[Datos origen](https://www.globalcarbonproject.org/carbonbudget/21/data.htm)_ 
 - Greenhouse gas emissions (including methane and nitrous oxide) by sector (CAIT): _[Datos origen](https://www.climatewatchdata.org/data-explorer/historical-emissions)_ 
 - CO2 dataset

Adicionalmente, para construir las varibales _per capita_ y _per GDP_, se usaron los siguientes datasets(si se desea ahondar en los ETL's de cada fuente, se pueden consultar en la documentación del dataset co2-data):

 - Population (Our World in Data based on a number of different sources).
 - GDP (University of Groningen GGDC's Maddison Project Database, Bolt and van Zanden, 2020). _[Datos origen](https://www.rug.nl/ggdc/historicaldevelopment/maddison/releases/maddison-project-database-2020)_
